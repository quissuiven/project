```python
# beds + baths + area + $psf

bed = [val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc Listing__summaryText__1QR5z'}) if 'Bed' in val.text]
bath = [val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc Listing__summaryText__1QR5z'}) if 'Bath' in val.text]
area = [val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc Listing__summaryText__1QR5z'}) if 'sqft' in val.text]
price_psf = [val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc Listing__summaryText__1QR5z'}) if 'psf' in val.text]
subline = [val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc'}) if 'Room' in val.text]
nearest_mrt = [val.text for val in soup.findAll('p',{'class':'NearestMrt__title__iVNyM'})]
amenity = [val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc Listing__amenityLabel__CQblY'})]
description = soup.findAll('pre',{'class':'Paragraph__text__lEhtw'})[0].text.replace("\n"," ")
completion_year = [val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc'}) if 'Year of Completion' in val]
price = soup.findAll('h3',{'class':'Heading__heading__2ncUp'})[0].text
furnishing = [val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc'}) if 'furnish' in val.text]

bed,bath,area,price_psf,subline,nearest_mrt,amenity,description,completion_year,price,furnishing
```

```python
url = pd.read_csv('listing_url_2.csv')

start_time = time.time()

bed = []
bath = []
area = []
price_psf = []
subline = []
nearest_mrt = []
amenity = []
description = []
completion_year = []
price = []
furnishing = []
address_copy = []
url_copy = []

# store website that is broken
broken_url=[]

chromedriver = "Users/swoo/Desktop/DSI/materials/classes/week-06/labs/python-webscraping_opentable-lab-master/chromedriver/chromedriver.exe"
os.environ["webdriver.chrome.driver"] = chromedriver
driver = webdriver.Chrome(executable_path="./chromedriver/chromedriver.exe")

# to adjust index of URLs that you want to scrap

start_index = 3000
end_index = None

for index,u in enumerate(url['url'][start_index:]):
    
    seconds = random.randint(3,6)
    
    url_ = 'https://www.99.co' + u
    
    driver.get(url_)
    sleep(seconds)
    
    response = requests.get(url_,headers={'User-Agent': 'Internet Explorer/11.706.17134.0'})
    html = driver.page_source
    soup = BeautifulSoup(html,'lxml')    
    
    if len(soup.findAll('h3',{'class':'Heading__heading__2ncUp'}))>0:
        
        print(index + start_index ,url_ ,seconds)
        
        bed.append([val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc Listing__summaryText__1QR5z'}) if 'Bed' in val.text])
        bath.append([val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc Listing__summaryText__1QR5z'}) if 'Bath' in val.text])
        area.append([val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc Listing__summaryText__1QR5z'}) if 'sqft' in val.text])
        price_psf.append([val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc Listing__summaryText__1QR5z'}) if 'psf' in val.text])
        subline.append([val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc'}) if 'Room' in val.text])
        nearest_mrt.append([val.text for val in soup.findAll('p',{'class':'NearestMrt__title__iVNyM'})])
        amenity.append([val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc Listing__amenityLabel__CQblY'})])
        
        try:
            description.append(soup.findAll('pre',{'class':'Paragraph__text__lEhtw'})[0].text.replace("\n"," "))
        except:
            description.append(None)
        
        completion_year.append([val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc'}) if 'Year of Completion' in val])
        price.append(soup.findAll('h3',{'class':'Heading__heading__2ncUp'})[0].text)
        furnishing.append([val.text for val in soup.findAll('p',{'class':'Text__text__x0JSc'}) if 'furnish' in val.text])
        address_copy.append(url['address'][index+start_index])
        url_copy.append(u)
    else: 
        print(url_+" is broken")
        broken_url.append(url_)

print("--- %s seconds ---" % (time.time() - start_time))

driver.close()
```

```python
# listing infor 2 - save it as csv

listing_details = list(zip(address_copy,url_copy,bed,bath,area,price_psf,subline,nearest_mrt,amenity,description,completion_year,price,furnishing))

col = ['address','url','beds','baths','area','price_psf','subline','nearest_mrt','amenities'
           ,'description','year_completion','price','furnishing']

pd.DataFrame(listing_details,columns = col).to_csv('all_listings_3.csv')
```
